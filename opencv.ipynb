# WORKING WITH VIDEO CAPTURES

```python
vdo_path = os.path.join('videos','cr7_bicycle_kick.gif')
vdo_path

```

```python
cap = cv2.VideoCapture(vdo_path)
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*'H264')

video_writer = cv2.VideoWriter(os.path.join('videos','gray_video.avi'),fourcc,fps,(width,height))

for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    
    ret ,frame = cap.read()
    
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow('CR7',gray)
    
    video_writer.write(gray)
    
    frame_rate = cap.get(cv2.CAP_PROP_FPS)
    delay = int(1000/frame_rate)
    
    if cv2.waitKey(delay) & 0xff == ord('q'):
        break
        
cap.release()
video_writer.release()
cv2.destroyAllWindows()


    
    
```
